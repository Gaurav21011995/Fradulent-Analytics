# Courier company Data File

```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

 Reading all the files

```python
order=pd.read_excel('Company X - Order Report.xlsx')
sku=pd.read_excel('Company X - SKU Master.xlsx')
invoice=pd.read_excel('Courier CompanyInvoice.xlsx')
rates=pd.read_excel('Courier Company - Rates.xlsx')
```

```python
order
```

We will rename the 'ExternOrderNo' column to 'Order ID' before merging them

```python
order.rename(columns={'ExternOrderNo':'Order ID'},inplace=True)
```

```python
order
```

We will merge the order and invoice data

```python
order_in=pd.merge(order,invoice)
```

```python
df=pd.merge(order_in,sku)
```

```python
df
```

Changing the weight from 'g' into 'kg'

```python
df['Weight_kg']=df['Weight (g)']/1000
```

```python
df
```

Creating New column naming Weight_slab and mutiply the Weight with Order column

```python
df['Weight_Slab']=df['Weight_kg']*df['Order Qty']
```

```python
df
```

Replacing all the value with 500 gram in Weight slab column where the value is lower than 500 gram by running np.where condition

```python
df.Weight_Slab = np.where(df.Weight_Slab<0.5,.5,df.Weight_Slab)
```

```python
df
```

Replacing all the value with 1 kg in Weight slab column where the value is more that 500 gram and lower than 1 kg by running np.where with multiple condition

```python
df.Weight_Slab = np.where((df.Weight_Slab>0.5) & (df.Weight_Slab<1.0),1.0,df.Weight_Slab)
```

```python
df
```

Replacing all the value with 1.5 kg in Weight slab column where the value is more than 1 kg by running np.where condition

```python
df.Weight_Slab = np.where(df.Weight_Slab>1.0,1.5,df.Weight_Slab)
```

```python
df
```

Now run values count command to check where all the 3 condition's has been sucessfully applied 

```python
df.Weight_Slab.value_counts()
```

Created another column as Difference to check that if the person is correctly charged, over charged, under charged for zone b
